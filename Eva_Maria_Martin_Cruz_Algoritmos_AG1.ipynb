##Devoluci칩n de cambio por t칠cnica voraz

```python
def cambio_monedas(N, SM):
  SOLUCION = [0]*len(SM)   #SOLUCION = [0,0,0,0,..]
  ValorAcumulado = 0

  for i,valor in enumerate(SM):
    monedas =  (N-ValorAcumulado)//valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if ValorAcumulado == N:
      return SOLUCION


cambio_monedas(15,[25,10,5,1])
```

## N-Reinas por t칠cnica de vueta atr치s


```python
def escribe(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")


def es_prometedora(SOLUCION,etapa):
  #print(SOLUCION)
  #Si la soluci칩n tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " est치 " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True



def reinas(N, solucion=[], etapa=0):
  if len(solucion) == 0:
      solucion=[0 for i in range(N)]

  for i in range(1, N+1):
    solucion[etapa] = i

    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
        escribe(solucion)
        print()
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

reinas(8)
```

# Dado un conjunto de puntos se trata de encontrar los dos puntos m치s cercanos.                                                                            C1:[3403, 4537, 9089, 9746, 7259,21, 342, 1003]


Para encontrar los dos puntos m치s cercanos en un conjunto de puntos utilizando fuerza bruta, se puede fijar el primero comparar la distancia entre 칠l y todos los dem치s quedandome con el m칤nimo. 
en Python:

```python
# Lista de puntos
C1 = [3403, 4537, 9089, 9746, 7259, 21, 342, 1003]

# Inicializar la distancia m칤nima global a un valor muy grande
min_distance_global = 9089

# Inicializar variables para guardar los puntos m치s cercanos globales
point1_global = None
point2_global = None

# Recorrer cada punto
for i in range(len(C1) - 1):
    # Inicializar la distancia m칤nima local para el punto actual
    min_distance_local = float('inf')
    
    # Comparar con todos los puntos siguientes
    for j in range(i + 1, len(C1)):
        distance = abs(C1[i] - C1[j])
        if distance < min_distance_local:
            min_distance_local = distance
            point1_local = C1[i]
            point2_local = C1[j]
    
    # Comparar la distancia m칤nima local con la distancia m칤nima global
    if min_distance_local < min_distance_global:
        min_distance_global = min_distance_local
        point1_global = point1_local
        point2_global = point2_local

print(f"Los puntos m치s cercanos son {point1_global} y {point2_global} con una distancia de {min_distance_global}.")
```

La complejidad de este algoritmo es cuadratica por tener un buble interno y un bucle externo

Aplicamos DIVIDE Y VENCER츼S

```python
C1 = [3403, 4537, 9089, 9746, 7259, 21, 342, 1003]

def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)

def find_closest_pair(points):
    sorted_points = quicksort(points)
    min_distance = float('inf')
    closest_pair = None
    for i in range(len(sorted_points) - 1):
        for j in range(i + 1, len(sorted_points)):
            distance = abs(sorted_points[j] - sorted_points[i])
            if distance < min_distance:
                min_distance = distance
                closest_pair = (sorted_points[i], sorted_points[j])
    return closest_pair, min_distance

closest_pair, min_distance = find_closest_pair(C1)
print("Pares m치s cercanos:", closest_pair)
print("Distancia:", min_distance)
```

Algoritmo de Divide y Vencer치s
Ordenar los Puntos: Primero, ordenamos los puntos en la lista.
Luego dividimos la lista en dos mitades, encontramos los pares m치s cercanos en cada mitad recursivamente, y luego combinamos los resultados.
Al ejecutar este c칩digo, se ordenar치 la lista de puntos utilizando QuickSort, y luego se iterar치 sobre la lista ordenada para encontrar los dos puntos m치s cercanos. 
El resultado ser치 el par de puntos m치s cercanos y la distancia entre ellos.
Este algoritmo tien una complejidad temporal de 洧녝(洧녵log洧녵)

## Retrieval Augmented Generation (RAG) - DISTANCIA ENTRE PUNTOS EN EL ESPACIO VECTORIAL

Resumen:

Un RAG, acr칩nimo de 춺Retrieval Augmented Generation췉, representa una estrategia innovadora dentro del procesamiento de lenguaje natural. Se integra con los Grandes Modelos de Lenguaje (LLM, por sus siglas en ingl칠s), tales como los que usa ChatGPT internamente (GPT-3.5-turbo o GPT-4), con el objetivo de mejorar la calidad de la respuesta y reducir ciertos comportamientos no deseados, como las alucinaciones.

Estos sistemas combinan los conceptos de vectorizaci칩n y b칰squeda sem치ntica, junto con los LLMs para retroalimentar su conocimiento con informaci칩n externa que no se incluy칩 durante su fase de entrenamiento y que, por lo tanto, desconocen.

<img src="Captura3.png">

La idea intuitiva del funcionamiento de un RAG es la siguiente:

El usuario hace una pregunta. 
Transformamos la pregunta a un VECTOR. 
Esto nos va a permitir comparar nuestra pregunta con toda la informaci칩n que tenemos indexada en nuestra BASE DE DATOS VECTORIAL.

CALCULAMOS LAS DITANCIAS entre la pregunta y todos los vectores que tenemos en la base de datos. 

Seleccionamos, con una estrategia, algunos de los chunks y a침adimos todas esas piezas de informaci칩n dentro del prompt como contexto.
La estrategia m치s sencilla es basarse en seleccionar un n칰mero (K) de VECTORES M츼S PR칍XIMOS a la pregunta.

Se lo pasamos al LLM para que genere la respuesta en base a los contextos.
Es decir, el prompt contiene instrucciones + pregunta + contexto devuelto por el sistema de Retrieval. 

Por este motivo, la parte de 춺Augmentation췉 en las siglas del RAG, dado que estamos haciendo prompt augmentation.

El LLM nos ha generado una respuesta en base a la pregunta que hacemos y el contexto que le hemos pasado. 
Esta ser치 la respuesta que el usuario va a visualizar.

Es por eso que necesitamos un embedding y la BASE DE DATOS VECTORIAL.
